```python
num = 3620
event = sxs.load(f"SXS:BBH:{num}", ignore_deprecation = True) #auto_supersede = True
w= event.h
```

```python
location = event.location
```

```python
plt.plot(w.t, w.data.view(float))
reference_time = event.metadata.reference_time
plt.axvline(x = reference_time, color = "red", label = "Reference Time")
var = "Extrapolated"
plt.title(f"{var} waveform") #f-strings allow you to do fancy things like put a variable in a string
plt.xlabel(r"$(t_{\mathrm{corr}} - r_\ast)/M$") #raw strings (r-strings) let the computer ignore the backslashes in the string (not interpreted as escapes)
plt.ylabel(r"$r\, h^{(\ell,m)}/M$");
plt.legend()
```

```python
reference_index = w.index_closest_to(reference_time)
w_sliced = w[reference_index:]

plt.plot(w_sliced.t, w_sliced.data.view(float))
plt.title(f"Sliced {var} waveform")
plt.xlabel(r"$(t_{\mathrm{corr}} - r_\ast)/M$")
plt.ylabel(r"$r\, h^{(\ell,m)}/M$");
```

```python
Top_Modes = np.array([])
Mode_Max = np.ndarray(shape=(len(w_sliced.data.view(float)[0]),2))

i = 0
for i in range(len(w_sliced.data.view(float)[0])):
    for strain in w_sliced.data.view(float):
        if strain[i] > Mode_Max[i][0]:
            Mode_Max[i] = strain[i], i

top4 =  sorted(Mode_Max, key=lambda x: x[0], reverse=True)[:4]

print(top4)
    
        
    #Mode_Max[i] = np.max(mode)
```

```python
plt.plot(w[:, int(top4[0][1]): int(top4[0][1]+1)].t, w[:, int(top4[0][1]): int(top4[0][1])+1].abs, label = int(top4[0][1]))
plt.plot(w[:, int(top4[1][1]): int(top4[1][1]+1)].t, w[:, int(top4[1][1]): int(top4[1][1]+1)].abs, label = int(top4[1][1]))
plt.plot(w[:, int(top4[2][1]): int(top4[2][1]+1)].t, w[:, int(top4[2][1]): int(top4[2][1]+1)].abs, label = int(top4[2][1]))
plt.plot(w[:, int(top4[3][1]): int(top4[3][1]+1)].t, w[:, int(top4[3][1]): int(top4[3][1]+1)].abs, label = int(top4[3][1]))
plt.xlabel(r"$(t_{\mathrm{corr}} - r_\ast)/M$")
plt.ylabel(r"$\left| r\, h^{(\ell,m)}/M \right|$")
plt.title(f"Top 4 Modes of SXS:BBH:{num}")
plt.legend()
#The legend has the mode indices listed, not the (l,m) pair specifying the mode
#Here we have the absolute values of the strains listed for each mode, which is actually the mode's weight in the waveform decomposition into spherical
#harmonics
```
